# Collaborative Based

```python
ratings_books.groupby('User-ID').count()['Book-Rating']
```

```python
user_rating = ratings_books.groupby('User-ID').count()['Book-Rating']>200
```

```python
user_rating
user_rating[user_rating]
```

```python
good_user = user_rating[user_rating].index
```

```python
good_user
```

```python
filtered_rating = ratings_books[ratings_books['User-ID'].isin(good_user)]
```

```python
filtered_rating
```

```python
x = filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
famous_books = x[x].index
```

```python
famous_books
```

```python
filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
final_rating = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
pivot = final_rating.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pivot
```

```python
pivot.fillna(0,inplace=True)
```

```python
pivot
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score = cosine_similarity(pivot)
```

```python
similarity_score
```

```python
def recommend(bookName):
    index = np.where(pivot.index==bookName)[0][0]
    similar_items = sorted(list(enumerate(similarity_score[index])),key=lambda z:z[1],reverse=True)[1:6]
    
    for i in similar_items:
        print(pivot.index[i[0]])
```

# Now you can use recommendation system

```python
recommend('Harry Potter and the Prisoner of Azkaban (Book 3)')
```

```python
recommend('A Bend in the Road')
```
